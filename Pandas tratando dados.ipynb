## Criando novas variaveis

```python
dados_residencial2 = pd.read_csv('Dados residencial Tratamento 5 valores nulos.csv', sep = ';')
```

```python
dados_residencial2['Valor Bruto'] 
```

```python
dados_residencial2['Media m2']  = dados_residencial2['Valor Bruto'] / dados_residencial2['Area']
```

```python
dados_residencial2['Media m2'] = dados_residencial2['Media m2'].round(2)
```

```python
dados_residencial2
```
